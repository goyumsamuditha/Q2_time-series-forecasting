```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
with open('point_metrics.json', 'r') as f:
    points_metrics = json.load(f)
with open('garch_metrics.json', 'r') as f:
    garch_metrics = json.load(f)

print("Points Metrics:", points_metrics)
print("GARCH Metrics:", garch_metrics)
```

```python
sns.set_theme(style="whitegrid")
```

```python
# Summarize dataframes
points_df = pd.DataFrame.from_dict(points_metrics, orient='index')
garch_df = pd.DataFrame.from_dict(garch_metrics, orient='index')
print("Point Forecast Models (Price Prediction)\n")
display(points_df)
print("GARCH DataFrame:\n")
display(garch_df)
```
